**Popularity Based Recommender System**

```python
rating_with_name = rating.merge(book,on='ISBN')
```

```python
rating_with_name.shape
```

```python
rating_with_name.head()
```

```python
num_rating_df = rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns = {'Book-Rating':'num_rating'},inplace=True)
num_rating_df
```

```python
avg_rating_df = rating_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
num_rating_df.rename(columns = {'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_rating']>=250].sort_values('Book-Rating',ascending=False)
popular_df.head(50)
```

```python
popular_df = popular_df.merge(book,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','Book-Rating']]
```

```python
popular_df
```

**Collaborative filtering based recommender system**

```python
x = rating_with_name.groupby('User-ID').count()['Book-Rating']>200
book_lover_users =x[x].index
```

```python
filtered_rating = rating_with_name[rating_with_name['User-ID'].isin(book_lover_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_book = y[y].index
famous_book
```

```python
final_rating = filtered_rating[filtered_rating['Book-Title'].isin(famous_book)]
```

```python
final_rating
```

```python
final_table = final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
final_table.fillna(0,inplace = True)
final_table
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(final_table)
similarity_score
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
  #fetch index
  index = np.where(final_table.index==book_name)[0][0] # Use the book_name variable instead of the string 'book_name'
  similar_item = sorted(list(enumerate(similarity_score[index])),key = lambda x:x[1], reverse = True)[1:6]
  for i in similar_item:
    # Use .iloc to access the row by index
    print(final_table.index[i[0]])
```

```python
recommend('The Notebook')
```
